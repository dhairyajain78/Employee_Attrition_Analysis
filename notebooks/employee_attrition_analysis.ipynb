3. Analyse department wise turnover

```python
emp_in_dept = ds['Department'].value_counts()
emp_in_dept
```

```python
emp_attrition_dept= ds.groupby('Department')['Attrition'].value_counts()
emp_left_dept=emp_attrition_dept.xs("No",level="Attrition")
emp_left_dept
```

```python
total_emp_deptwise = ds.groupby('Department')['EmployeeCount'].sum()
total_emp_deptwise 
```

```python
turnover = (emp_left_dept*100)/total_emp_deptwise
```
