1. Analyze employee attrition patterns.

```python
pattern=ds.groupby('Gender')['Attrition'].value_counts()
pattern
```

2. Study salary impact on attrition

```python
ds.groupby('PercentSalaryHike')['Attrition'].value_counts()
```

3. Analyse department wise turnover

```python
emp_in_dept = ds['Department'].value_counts()
emp_in_dept
```

```python
emp_attrition_dept= ds.groupby('Department')['Attrition'].value_counts()

emp_left_dept=emp_attrition_dept.xs("Yes",level="Attrition")
emp_left_dept
```

```python
emp_attrition_dept
```

```python
total_emp_deptwise = ds.groupby('Department')['EmployeeCount'].sum()
total_emp_deptwise 
```

```python
turnover = (emp_left_dept/total_emp_deptwise)*100
```

```python
turnover
```

4 . Investigate overtimr effects

```python
ds.groupby('OverTime')['PercentSalaryHike'].value_counts()
```

5. Recommend retention strategies

```python
emp_remaining_dept = emp_attrition_dept.xs("No",level ="Attrition")
emp_remaining_dept
```

```python
retention_rate = (emp_remaining_dept/ total_emp_deptwise)*100
retention_rate
```

Expected Visualizations

Attrition count plots


```python
sns.countplot(x=ds['Attrition'])
plt.title("Employee Attrition Count")
plt.show()
```

Salary distribution histograms

```python
ds.groupby('JobRole')['MonthlyIncome'].mean()
```

```python
sns.violinplot(x=ds['JobRole'],y=ds['MonthlyIncome'])
plt.title("Job Role vs Monthly Income")
plt.xticks(rotation=90)
plt.show()
```

Department comparisons


```python
compare = ds.groupby('Department')['EmployeeCount'].sum()
compare
```

```python
sns.barplot(x=compare.index,y=compare.values, color='pink')
plt.title("Department wise Employee Count")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.show()
```
